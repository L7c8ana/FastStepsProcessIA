## Train set - used to learning and backpropagation

```python
print(IMGT.shape)
###### Split train,val,test sets
x_train, x_test, y_train, y_test = train_test_split(IMGT, CLS1k, test_size=0.1, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.1, random_state=42)
x_trainMN, x_valMN, y_trainMN, y_valMN = train_test_split(x_trainMN, y_trainMN, test_size=0.4, random_state=42)
print(x_trainMN.shape)
```

## 7 - Create simple model sequential with tensorflow Dense layers

### Tensorflow is framework in Python to Machine Learning

```python
###### Making Sequential layers model function
def MODEL(input, output):
  model = tf.keras.Sequential([
      tf.keras.layers.Flatten(input_shape=(input.shape[1], input.shape[2])),
      tf.keras.layers.Dense(256, activation='relu'),
      tf.keras.layers.Dropout(0.1),
      tf.keras.layers.Dense(16, activation='relu'),
      tf.keras.layers.Dropout(0.2),
      tf.keras.layers.Dense(output.shape[1], activation='sigmoid')
  ])
  ###### compile model with Adam optimizer and categorical_crossentropy loss
  model.compile(optimizer=tf.keras.optimizers.Adam(),
                      loss='categorical_crossentropy',
                                    metrics=['accuracy'])
  return model
print(x_trainMN.shape)
model = MODEL(x_train,y_train)
modelMN = MODEL(x_trainMN,y_trainMN)


###### show summary model
print(model.summary())
```

# Plot ROC
### ROC is a plot showing the performance of a classification model at all classification thresholds. This curve plots two parameters: True Positive Rate. False Positive Rate.

#### First predict in test set after make plot roc from predictions.

```python
predsMN = modelMN.predict(x_testMN)
for i in range(y_testMN.shape[1]):
  fprMN, tprMN, thresholdMN = metrics.roc_curve(y_testMN[:,i], predsMN[:,i])
  roc_aucMN = metrics.auc(fprMN, tprMN)

  plt.title('Receiver Operating Characteristic')
  plt.plot(fprMN, tprMN, color='C'+str(i), label = 'cls '+str(i)+' AUC = %0.2f' % roc_aucMN)
plt.legend(loc = 'lower right')
#plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()

###### predict model using test set
preds = model.predict(x_test)
fpr, tpr, threshold = metrics.roc_curve(y_test[:,0], preds[:,0])
roc_auc = metrics.auc(fpr, tpr)
plt.figure()
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
#plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()


```
